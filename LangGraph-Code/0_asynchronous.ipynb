# Synchronous programming

```python
import time
```

```python

def fetch_weather():
    print("Fetching weather data...")
    time.sleep(4)  # Simulate a network delay
    print("Weather data fetched.")


def fetch_news():
    print("Fetching news data...")
    time.sleep(2)  # Simulate a network delay
    print("News data fetched.")



def main():
    start_time = time.time()

    fetch_weather()
    fetch_news()

    end_time = time.time()
    print(f"Total time taken: {end_time - start_time} seconds")




main()
```

# Asynchronous programming

#### asyncio is Python's built-in library for writing concurrent programs using asynchronous programming. It lets a single thread handle many tasks efficiently, especially tasks that spend time waiting (network requests, APIs, databases, file I/O, etc.).

#### It does not make your code faster by using more CPU cores. Instead, it prevents your program from sitting idle while waiting for I/O.

```python
import asyncio
import time
```

```python
async def fetch_weather():
    print("Fetching weather data...")
    await asyncio.sleep(4)  # Simulate a network delay
    print("Weather data fetched.")


async def fetch_news():
    print("Fetching news data...")
    await asyncio.sleep(2)  # Simulate a network delay
    print("News data fetched.")



async def main():
    start_time = time.time()

    await asyncio.gather(fetch_weather(), fetch_news())

    end_time = time.time()
    print(f"Total time taken: {end_time - start_time} seconds")



await main()
```
